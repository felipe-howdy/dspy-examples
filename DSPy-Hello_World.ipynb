2️⃣ Import and configure the LLM

```python
import dspy

# connect to your LLM
lm = dspy.LM(
    "openai/gpt-4.1-2025-04-14",
    api_key="INSERT YOUR API KEY HERE"  # ← replace with your key
)

dspy.configure(lm=lm)
```

3️⃣ Define the module (signature)

```python
# Define a simple QA module
qa = dspy.Predict("question -> answer")
```

4️⃣ Run it

```python
response = qa(question="What is context rot?")
print(response.answer)
```
